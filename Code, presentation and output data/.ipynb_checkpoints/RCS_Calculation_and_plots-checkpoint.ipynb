```python
dates_considered = ['08-13-2021','09-06-2021','09-12-2021','09-30-2021','10-12-2021','10-24-2021','11-29-2021']

CR_RCS_BG_C=[0.0896,0.1021,0.8619,0.8758,0.1522,0.0998,0.1605]
CR_BG_C=[0.198,0.248,0.2832,0.4022,0.3793,0.2529,0.4988]
CR_NO_P=[120,120,132,132,110,120,110]


ECR1_RCS_BG_C=[1.8183,1.8924,1.9018,1.7616,2.3099,2.1124,2.2535]
ECR1_BG_C=[0.8545,0.9977,0.5897,0.7116,0.8302,0.4948,0.4522]
ECR1_NO_P=[168,182,195,195,168,180,168]


ECR2_RCS_BG_C=[1.1215,1.4358,1.3754,1.3724,2.0148,1.2956,1.7813]
ECR2_BG_C=[0.7216,0.6065,0.6061,0.6335,0.8055,0.5959,0.5692]
ECR2_NO_P=[132,132,130,130,130,143,120]
```

```python
CR_CAL_RCS = []
for i in range(len(dates_considered)):
    rcs = 10*np.log10(100*CR_NO_P[i]*(CR_RCS_BG_C[i]-CR_BG_C[i]))
    CR_CAL_RCS.append(rcs)
    
ECR1_CAL_RCS = []
for j in range(len(dates_considered)):
    rcs = 10*np.log10(100*ECR1_NO_P[j]*(ECR1_RCS_BG_C[j]-ECR1_BG_C[j]))
    ECR1_CAL_RCS.append(rcs)

ECR2_CAL_RCS = []
for k in range(len(dates_considered)):
    rcs = 10*np.log10(100*CR_NO_P[k]*(ECR2_RCS_BG_C[k]-ECR2_BG_C[k]))
    ECR2_CAL_RCS.append(rcs)

```

```python
CR_CAL_RCS = list(np.around(np.array(CR_CAL_RCS),2))
ECR1_CAL_RCS = list(np.around(np.array(ECR1_CAL_RCS),2))
ECR2_CAL_RCS=list(np.around(np.array(ECR2_CAL_RCS),2))

print(CR_CAL_RCS)
print(ECR1_CAL_RCS)
print(ECR2_CAL_RCS)
```

```python
plt.ylim(36,45)
plt.xlabel('Date',fontweight="bold")
plt.xticks(rotation=90)
plt.ylabel('Amplitude in dB',fontweight="bold")
plt.plot(dates_considered,CR_CAL_RCS,"-o",label='Amplitude Values of CR')
plt.plot(dates_considered,ECR1_CAL_RCS,"-o",label='Amplitude Values of ECR 1')
plt.plot(dates_considered,ECR2_CAL_RCS,"-o",label='Amplitude Values of ECR 2')
plt.legend(loc='upper left',bbox_to_anchor=(1, 0.25))
plt.title("Sigma0 Values (dB) of Point Targets in Oberfaffenhofen",fontweight="bold")
plt.savefig('RCSx.jpeg')
```
